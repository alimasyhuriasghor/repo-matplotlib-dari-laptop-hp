```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {
    'A':109438,
    'B':103569,
    'C':112214,
    'D':112591,
    'E':100934,
    'F':103660,
    'G':137351,
    'H':123381,
    'I':135841,
    'J':104437
}
df = pd.Series(data)
df
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

# Simple OO Style Bar Horizontal Plot

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.grid()
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()

ax.bar(items,count)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Simple Bar Vertikal plot')

plt.show()
```

# Pengaturan Tick Labels

```python
fig, ax = plt.subplots()

ax.barh(items,count)

labels = ax.get_xticklabels()

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Simple Bar Horizontal plot')

plt.setp(labels,rotation=45
         ,horizontalalignment='right')

fig.tight_layout()
plt.show()
```

# Pengaturan Format Pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x,pos):
    return f"{int(x/1000)}K"
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Plot')

fig.tight_layout()
plt.show()
```

# Pengaturan label Pada Sumbu (axis) Dan Judul (title)

```python
fig, ax = plt.subplots()

ax.barh(items,count)
ax.set(title='contoh plot',xlabel='jumlah',ylabel='count')

plt.show()
```

# Penambahan Garis (vertical/horizontal line) pada plot

```python
fig, ax = plt.subplots()

ax.barh(items,count)
ax.axvline(np.mean(count),ls='--',c='r')

ax.set(title='contoh plot',xlabel='jumlah',ylabel='count')

plt.show()
```

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.axvline(np.mean(count),linestyle='-',color='r')

ax.set_title('Nyimpen Gambar')
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

fig.tight_layout()
fig.savefig('hasil.jpg',transparent=False,dpi=80)
```
